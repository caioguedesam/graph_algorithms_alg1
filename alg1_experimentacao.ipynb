```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('output/output.csv')
df
```

```python
df_means = df.groupby(['V+A']).agg({'Tempo':'mean'})
df_means
```

```python
plt.figure(figsize=(10,5))
#plt.scatter(df_means.index.values, df_means['Tempo'])
plt.plot(df_means.index.values, df_means['Tempo'], lw = 2, color='goldenrod')
plt.xlabel('Vértices (alunos) + Arestas (relações de comando)')
plt.ylabel('Média de tempo (em nanossegundos)')
plt.show()
```

```python
df_sdv = df.groupby(['V+A']).agg({'Tempo':'std'})
df_sdv
```

```python
plt.figure(figsize=(10,5))
#plt.scatter(df_means.index.values, df_means['Tempo'])
plt.plot(df_sdv.index.values, df_sdv['Tempo'], lw = 2, color='teal')
plt.xlabel('Vértices (alunos) + Arestas (relações de comando)')
plt.ylabel('Desvio Padrão do tempo (em nanossegundos)')
plt.show()
```
